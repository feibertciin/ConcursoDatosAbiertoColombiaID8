Evaluación de confiabilidad, curvas ROC y métricas de desempeño de los modelos.

```python
import sys
import os

# Buscar dinámicamente la raíz del proyecto que contiene la carpeta 'src'
current_dir = os.getcwd()
found = False
while current_dir and current_dir != os.path.dirname(current_dir):
    if os.path.exists(os.path.join(current_dir, 'src')):
        sys.path.insert(0, current_dir)
        found = True
        print(f"✅ Raíz del proyecto encontrada en: {current_dir}")
        break
    current_dir = os.path.dirname(current_dir)

if not found:
    print("⚠️ Advertencia: No se pudo localizar la carpeta 'src' subiendo en el árbol de directorios.")

import pandas as pd
from src.infrastructure.database.connection import SessionLocal
from src.infrastructure.database.models import ModelMetadata

db = SessionLocal()
models = db.query(ModelMetadata).all()
db.close()

print("Métricas de modelos en producción registrados en base de datos:")
for m in models:
    print(f"- {m.model_name}: Accuracy={m.accuracy:.4f}, F1-Score={m.f1_score:.4f}")
```
