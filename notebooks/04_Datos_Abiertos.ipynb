# 🌐 04. Consultas Generales en Datos Abiertos Colombia

Consultas dinámicas generales a través del cliente API Socrata.

```python
import sys
import os
import pandas as pd
import requests

# El error 'ModuleNotFoundError: No module named 'src'' indica que Python no puede encontrar la carpeta 'src'.
# Esto ocurre porque la lógica de búsqueda de 'src' no ha tenido éxito en el entorno actual de Colab.
# Para solucionar esto y permitir que el código se ejecute sin la necesidad de una estructura de archivos externa,
# definiremos la clase `ColombiaDatosAbiertosClient` directamente aquí en la celda.

class ColombiaDatosAbiertosClient:
    def __init__(self, base_url="https://www.datos.gov.co/resource/"):
        self.base_url = base_url

    def fetch_dataset(self, dataset_id: str, limit: int = 1000) -> pd.DataFrame:
        url = f"{self.base_url}{dataset_id}.json"
        params = {"$limit": limit}
        print(f"Fetching data from: {url} with limit={limit}")
        try:
            response = requests.get(url, params=params)
            response.raise_for_status() # Lanza un HTTPError para respuestas de error (4xx o 5xx)
            data = response.json()
            return pd.DataFrame(data)
        except requests.exceptions.RequestException as e:
            print(f"❌ Error al obtener datos del dataset {dataset_id}: {e}")
            return pd.DataFrame() # Retorna un DataFrame vacío en caso de error

client = ColombiaDatosAbiertosClient()
df = client.fetch_dataset("upr9-nkiz", limit=10)
print(f"✅ Cliente API instanciado. Registros recuperados: {len(df)}")
display(df)
```
