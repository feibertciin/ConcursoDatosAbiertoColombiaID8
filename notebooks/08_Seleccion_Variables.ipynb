# 🎯 08. Selección de Variables Críticas

Evaluación de la importancia de las variables usando estimadores RandomForest.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import joblib

df = pd.read_csv("data/processed/processed_students.csv")
preprocessor = joblib.load("models/preprocessor.pkl")

X = preprocessor.transform(df)
y = df["desertor"]

forest = RandomForestClassifier(n_estimators=50, random_state=42)
forest.fit(X, y)

importances = forest.feature_importances_
print(f"✅ Variables evaluadas. Importancia media: {np.mean(importances):.4f}")
```
