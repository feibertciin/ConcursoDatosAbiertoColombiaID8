# ⚙️ 05. Pipeline de Extracción, Transformación y Carga (ETL)

Este cuaderno unifica SNIES (Académicos), DANE (Socioeconómicos) y OLE (Mercado laboral).

```python
import sys
import os

# Buscar dinámicamente la raíz del proyecto que contiene la carpeta 'src'
current_dir = os.getcwd()
found = False
while current_dir and current_dir != os.path.dirname(current_dir):
    if os.path.exists(os.path.join(current_dir, 'src')):
        sys.path.insert(0, current_dir)
        found = True
        print(f"✅ Raíz del proyecto encontrada en: {current_dir}")
        break
    current_dir = os.path.dirname(current_dir)

if not found:
    print("⚠️ Advertencia: No se pudo localizar la carpeta 'src' subiendo en el árbol de directorios.")

import pandas as pd
from src.application.etl import ETLPipeline

pipeline = ETLPipeline()
students, curriculum = pipeline.run_etl()

print("✅ Pipeline ETL ejecutado satisfactoriamente.")
print(f"Dimensión Estudiantes: {students.shape}")
print(f"Dimensión Pertinencia Curricular: {curriculum.shape}")
display(students.head(3))
```
