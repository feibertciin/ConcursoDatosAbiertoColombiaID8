# 🔍 06. Análisis Exploratorio de Datos (EDA)

Visualización de correlaciones, histogramas de deserción y promedios académicos por estrato.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/processed/processed_students.csv")
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="promedio_acumulado", hue="desertor", multiple="stack", palette="crest")
plt.title("Distribución del Promedio Académico vs Deserción")
plt.grid(True)
plt.show()
```
