# 👥 02. Descarga e Integración de Variables Socioeconómicas (DANE)

Este cuaderno consume indicadores socioeconómicos e ingresos familiares oficiales agregados por departamento en Colombia desde el portal de Datos Abiertos.

```python
import pandas as pd
import requests

# Consumo de dataset socioeconómico real en datos.gov.co (ej: estrato e ingresos por departamento)
url = "https://www.datos.gov.co/resource/ddau-8cy9.json" # Usando el principal del concurso
params = {
    "$select": "departamento_origen,estrato_socioeconomico,ingreso_familiar",
    "$limit": 500
}
response = requests.get(url, params=params)
if response.status_code == 200:
    df = pd.DataFrame(response.json())
    print("✅ Datos cargados del DANE/Socioeconómicos:")
    display(df.head())
else:
    print(f"Error: {response.status_code}")
```
