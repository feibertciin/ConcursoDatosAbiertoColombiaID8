# 👥 02. Descarga e Integración de Variables Socioeconómicas (DANE/Estrato)

Este cuaderno consume indicadores socioeconómicos e ingresos familiares oficiales agregados por departamento en Colombia desde el portal de Datos Abiertos.

```python
import pandas as pd
import requests

# Consumo de dataset socioeconómico real en datos.gov.co (ej: estudiantes inscritos por estrato)
url = "https://www.datos.gov.co/resource/36dj-k2a3.json"
params = {
    "$limit": 500
}
response = requests.get(url, params=params)
if response.status_code == 200:
    df = pd.DataFrame(response.json())
    print("✅ Datos cargados de Datos Abiertos (Socioeconómicos por Estrato):")
    display(df.head())
else:
    print(f"❌ Error: {response.status_code} - {response.text}")
```
