# 📋 16. Analítica Prescriptiva Dinámica

Uso del cliente API de Datos Abiertos de Colombia para alimentar el motor de recomendaciones prescriptivas regionales.

```python
import sys
import os

# Buscar dinámicamente la raíz del proyecto que contiene la carpeta 'src'
current_dir = os.getcwd()
found = False
while current_dir and current_dir != os.path.dirname(current_dir):
    if os.path.exists(os.path.join(current_dir, 'src')):
        sys.path.insert(0, current_dir)
        found = True
        print(f"✅ Raíz del proyecto encontrada en: {current_dir}")
        break
    current_dir = os.path.dirname(current_dir)

if not found:
    print("⚠️ Advertencia: No se pudo localizar la carpeta 'src' subiendo en el árbol de directorios.")

import pandas as pd
from src.infrastructure.external_api.colombia_api import ColombiaDatosAbiertosClient

client = ColombiaDatosAbiertosClient()
df = client.get_prescriptive_programs_data()

print("✅ Recomendaciones prescriptivas generadas a partir del portal Datos Abiertos:")
display(df[['departamento', 'tasa_desercion_historica', 'recomendacion_prescriptiva']].head())
```
