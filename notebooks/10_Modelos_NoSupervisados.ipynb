# 🧩 10. Agrupamiento No Supervisado (Clustering)

Segmentación de estudiantes vulnerables usando el algoritmo K-Means.

```python
from sklearn.cluster import KMeans
import pandas as pd
import joblib

df = pd.read_csv("data/processed/processed_students.csv")
preprocessor = joblib.load("models/preprocessor.pkl")

X = preprocessor.transform(df)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

df['cluster'] = clusters
print("✅ Estudiantes segmentados en 3 grupos (clusters):")
print(df.groupby('cluster')['promedio_acumulado'].mean())
```
