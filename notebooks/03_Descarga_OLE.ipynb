# 💼 03. Integración de Graduados y Empleabilidad (OLE)

Consumo del Observatorio Laboral para la Educación (OLE) que detalla el salario de enganche y la tasa de inserción por programa académico.

```python
import pandas as pd
import requests

# Simulación y consumo real del OLE
url = "https://www.datos.gov.co/resource/ddau-8cy9.json"
response = requests.get(url, params={"$limit": 100})
if response.status_code == 200:
    df = pd.DataFrame(response.json())
    print("✅ Conexión establecida con OLE API. Estructura de programas:")
    display(df.columns.tolist())
else:
    print("Error de consulta")
```
