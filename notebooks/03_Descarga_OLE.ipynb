# 💼 03. Integración de Graduados y Empleabilidad (OLE)

Consumo del portal de Datos Abiertos sobre programas de educación superior oficial para cruce de inserción laboral y salarios.

```python
import pandas as pd
import requests

# Consumo real del dataset de Programas de Educación Superior
url = "https://www.datos.gov.co/resource/upr9-nkiz.json"
response = requests.get(url, params={"$limit": 100})
if response.status_code == 200:
    df = pd.DataFrame(response.json())
    print("✅ Conexión establecida con la API del MEN. Estructura de programas:")
    display(df.columns.tolist())
    display(df.head())
else:
    print(f"❌ Error: {response.status_code} - {response.text}")
```
