# 🛠️ 07. Ingeniería de Características (Feature Engineering)

Preprocesamiento de datos: escalamiento numérico y codificación de variables categóricas.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

df = pd.read_csv("data/processed/processed_students.csv")
cat_cols = ['genero', 'departamento_origen', 'tipo_colegio', 'programa_academico']
num_cols = ['edad', 'puntaje_saber11', 'promedio_acumulado', 'materias_reprobadas', 'ingreso_familiar']

preprocessor = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
    ('num', StandardScaler(), num_cols)
])

X_proc = preprocessor.fit_transform(df)
print(f"✅ Matriz preprocesada con éxito. Nuevas columnas: {X_proc.shape[1]}")
```
